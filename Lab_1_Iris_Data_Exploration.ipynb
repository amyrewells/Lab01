## 3. Explore individual attributes

Investigate one or more attributes on their own.

Consider typical values, variability, the shape of the distribution, possible unusual observations, and whether the results change when the species are considered separately.

```python
# get spepal and petal length mean
print("Mean sepal length:", iris['sepal_length_cm'].mean())
print("Mean petal length:", iris['petal_length_cm'].mean())

# get sepal and petal length median
print("Median sepal length:", iris['sepal_length_cm'].median())
print("Median petal length:", iris['petal_length_cm'].median())

# get spepal and petal length mode
print("Mode sepal length:", iris['sepal_length_cm'].mode()[0])
print("Mode petal length:", iris['petal_length_cm'].mode()[0])

# get spepal and petal length standard deviation
print("Standard deviation sepal length:", iris['sepal_length_cm'].std())
print("Standard deviation petal length:", iris['petal_length_cm'].std())

# plot histograms of sepal and petal length
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(iris['sepal_length_cm'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.hist(iris['petal_length_cm'], bins=20, color='orange', alpha=0.7)
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

```

## 4. Explore relationships between attributes

Investigate relationships between two or more numeric attributes.

Consider whether measurements increase together, whether the species form visible groups, whether relationships look different for different species, and whether some pairs of attributes are more useful than others.

```python
# import scipy
%pip install scipy
import scipy.stats as stats
print("")

# get correlation between sepal length and petal length
correlation = iris['sepal_length_cm'].corr(iris['petal_length_cm'])
print("Correlation between sepal length and petal length:", correlation)

# get p vallue for correlation
corr, p_value = stats.pearsonr(iris['sepal_length_cm'], iris['petal_length_cm'])
print("P-value for correlation:", p_value)
```

## 5. Compare the species

Explore how the measurements differ among the three Iris species.

Consider:

- Which measurements differ most visibly by species?
- Which species appear easiest or hardest to distinguish?
- Where is there substantial overlap?
- Which summaries or visualizations help answer these questions?

```python
# subset data by species
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']

# get correlation between sepal length and petal length for each species
setosa_corr = setosa['sepal_length_cm'].corr(setosa['petal_length_cm'])
versicolor_corr = versicolor['sepal_length_cm'].corr(versicolor['petal_length_cm'])
virginica_corr = virginica['sepal_length_cm'].corr(virginica['petal_length_cm'])
print("Correlation between sepal length and petal length for setosa:", setosa_corr)
print("Correlation between sepal length and petal length for versicolor:", versicolor_corr)
print("Correlation between sepal length and petal length for virginica:", virginica_corr)

# get p value for correlation between sepal length and petal length for each species
setosa_corr, setosa_p_value = stats.pearsonr(setosa['sepal_length_cm'], setosa['petal_length_cm'])
versicolor_corr, versicolor_p_value = stats.pearsonr(versicolor['sepal_length_cm'], versicolor['petal_length_cm'])
virginica_corr, virginica_p_value = stats.pearsonr(virginica['sepal_length_cm'], virginica['petal_length_cm'])
print("P-value for correlation between sepal length and petal length for setosa:", setosa_p_value)
print("P-value for correlation between sepal length and petal length for versicolor:", versicolor_p_value)
print("P-value for correlation between sepal length and petal length for virginica:", virginica_p_value)
```

## 6. Choose useful visualizations

Try visualizations that match the questions you are asking.

```python
# Plot petal length vs sepal length
plt.figure(figsize=(8, 6))
plt.scatter(iris['sepal_length_cm'], iris['petal_length_cm'], color='green', alpha=0.7)
plt.title('Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

# add a line of best fit
slope, intercept, r_value, p_value, std_err = stats.linregress(iris['sepal_length_cm'], iris['petal_length_cm'])
line = slope * iris['sepal_length_cm'] + intercept
plt.plot(iris['sepal_length_cm'], line, color='red', linewidth=2)

plt.show()

%pip install seaborn
import seaborn as sns

# plot speal length vs petal length by species with lines of best fit
plt.figure(figsize=(8, 6))
sns.lmplot(x='sepal_length_cm', y='petal_length_cm', hue='species', data=iris, aspect=1.5)
plt.title('Sepal Length vs Petal Length with Lines of Best Fit')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
